2d plotting

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
N=100000
x=np.random.normal(size=N)
y=np.random.normal(size=N)
hist,xedges,yedges=np.histogram2d(x,y,bins=25)
print(hist.shape)
```

```python
plt.imshow(hist)
```

```python
plt.imshow(hist,extent=[xedges[0],xedges[-1],yedges[0],yedges[-1]])
```

```python
hist2,xedges2,yedges2=np.histogram2d(x,y,bins=25,range=[[-2,2],[-2,2]])
plt.imshow(hist2,extent=[xedges2[0],xedges2[-1],yedges2[0],yedges2[-1]])
```

```python
plt.imshow(hist2,cmap='plasma',extent=[xedges2[0],xedges2[-1],yedges2[0],yedges2[-1]])
```

```python
plt.imshow(hist2,cmap='coolwarm',extent=[xedges2[0],xedges2[-1],yedges2[0],yedges2[-1]])
```

```python
plt.imshow(hist2,cmap='brg',extent=[xedges2[0],xedges2[-1],yedges2[0],yedges2[-1]])
plt.colorbar()
```

```python
#adding different elements of the vector to itself
x=0.5*(xedges2[1:]+xedges2[0:-1])
y=0.5*(yedges2[1:]+yedges2[0:-1])
plt.contour(x,y,hist2)
```

```python
f,ax=plt.subplots(figsize=(4,4))
cs=ax.contour(hist2,[100,200,300,400],extent=[xedges2[0],xedges2[-1],yedges2[0],yedges2[-1]])
ax.clabel(cs,inline=1,fontsize=10)
```

```python
f,ax=plt.subplots(figsize=(4,4))
cs=ax.contour(hist2,[100,200,300,400],colors=['b','c','r','m'],extent=[xedges2[0],xedges2[-1],yedges2[0],yedges2[-1]])
ax.clabel(cs,inline=1,fontsize=10,fmt='%d')
```
