```python
#height of the ball after being dropped
g=9.81
h=float(input("Enter the height of the tower: "))
t=float(input("Enter how long the ball has been falling: "))
H=h-0.5*g*t**2
if(H>0 and t==0):
    print("The ball has not been dropped yet.")
elif(H>0):
    print("The height of the ball is:",H)
else:
    print("The ball has already hit the ground.")
```

```python
#Catalan numbers printing the first 10
n=0
C=1
L=[]
while(n < 10):
    C = (4*n+2)/(n+2)*C
    n = n+1
    L.append(C)
print(L)
```

```python
#plotting 2d arrays
#(-10,10) is the range and 100 is how many spaces in between
x=np.linspace(-10,10,100)
y=np.linspace(-10,10,100)
r=np.zeros((100,100))
k=1
E0=1
for i in range(len(x)):
    for j in range(len(y)):
        r[i][j]=np.sqrt(x[i]**2+y[j]**2)    #slicing r by i and j
plt.imshow(np.sin(k*r))
```
